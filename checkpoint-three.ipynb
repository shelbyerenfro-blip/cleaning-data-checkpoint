## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

I can remove the User ID column since it serves the same function as my index. I am also going to remove Reaction Time and AI Predicted Score, as I don't need these for my analysis. 
I don't know if this is the right place for this, but the idea came to me at this step so I'm adding it here. My dataset is really big, so I'm going add an age group column and categorize participants into Young Adult, Adult, and Older Adult. This will allow me to break down the data, and focus on the age group that interests me most (Adults). 

```python
hcp_edited = hcp_df.drop(columns=["User_ID", "Reaction_Time", "AI_Predicted_Score"])
```

```python
hcp_edited['Age_Group'] = hcp_edited.apply(
    lambda row: (
        'Young Adult' if row['Age'] < 25
        else 'Adult' if row['Age'] >= 25 and row['Age'] <= 40
        else 'Older Adult'
    ), axis = 1
)
```
